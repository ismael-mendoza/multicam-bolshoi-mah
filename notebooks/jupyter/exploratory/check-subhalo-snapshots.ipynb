```python
import numpy as np
import json
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy.io import ascii
```

```python
dmcat = ascii.read("../../../output/output_m14/dm_cat.csv", format='csv')
# t = ascii.read("../../../output/output_m14/subhaloes.csv", format='csv')
t = ascii.read('../../../../../Downloads/subhaloes.csv')

with open("../../../output/output_m14/z_map.json") as jp:
    z_map = json.load(jp)
```

```python
scales = list(z_map.values())

fig, axes = plt.subplots(3, 3, figsize=(11, 11))
axes = axes.flatten()
for i in range(9):
    ax = axes[i]
    assert dmcat['id'][i] == t['id'][i]
    f_subs = np.array([t[f'f_sub_a{j}'][i] for j in z_map])
    ax.plot(scales, f_subs, 'r-')
    ax.set_xlabel('scale', size=12)
    ax.set_ylabel('f_sub', size=12)
    ax.set_title(f"log10 mass: {np.log10(dmcat['mvir'][i]):.2f}")
plt.tight_layout()
```

```python
scales = list(z_map.values())

fig, axes = plt.subplots(3, 3, figsize=(11, 11))
axes = axes.flatten()
for i in range(9):
    ax = axes[i]
    assert dmcat['id'][i] == t['id'][i]
    m2s = [t[f'm2_a{j}'][i] for j in z_map]
    ax.plot(scales, m2s, 'ro')
    ax.set_xlabel('scale', size=12)
    ax.set_ylabel('m2', size=12)
    ax.set_title(f"log10 mass: {np.log10(dmcat['mvir'][i]):.2f}")
plt.tight_layout()
```

## Check f_sub > 1 

```python
f_subs = np.array([t[f'f_sub_a{j}'][1] for j in z_map])
scales = np.array(scales)
weird_indices = np.where(f_subs > 1)[0].astype(int)
scale = scales[weird_indices][0]
print(scale)
```
